```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('FINAL_DATA.CSV', sep =",", index_col = "MESS_DATUM")
df
```

```python
start = '2018-01-01'
end = '2022-12-31'
df_5_years = df.loc[start:end]
df_5_years
```

```python
if pd.api.types.is_datetime64_any_dtype(df_5_years.index):
    print('yes')
else:
    print('No')
```

```python
pd.options.mode.chained_assignment = None
df_5_years.index = pd.to_datetime(df_5_years.index)
df_5_years['Year'] = df_5_years.index.year
df_5_years['Month'] = df_5_years.index.month
#df_5_years['Total_precip'] = df_5_years['1590']+df_5_years['5733']+df_5_years['13670']
df_5_years
```

```python
df_5_years['Total_Precipitation'] = df_5_years[['1590', '5733', '13670']].sum(axis=1)
df_5_years.head()
```

```python
#Monthly = df_5_years.groupby(['Year','Month'])['Total_Precipitation'].idxmax()
#df_Monthly = df_5_years.loc[Monthly]
#df_Monthly
```

```python
#most = df_Monthly.groupby(['Year'])['Total_Precipitation'].idxmax()
#df_most = df_Monthly.loc[most]
#df_most
```

```python
monthly_accumulated = df_5_years['Total_Precipitation'].resample('M').sum()
df_5_years['Monthly_Accumulated'] = monthly_accumulated
df_5_years.head()
```

```python
Monthly = df_5_years.groupby(['Year','Month'])['Monthly_Accumulated'].idxmax()
df_Monthly = df_5_years.loc[Monthly]
df_Monthly.head()
```

```python
most_precip_month = df_Monthly.groupby(['Year'])['Monthly_Accumulated'].idxmax()
df_most_precip_month = df_Monthly.loc[most_precip_month]
df_most_precip_month
```

```python
df_match = pd.DataFrame()
for year, month in zip(df_most_precip_month['Year'], df_most_precip_month['Month']):
    mask = (df_5_years.index.year == year) & (df_5_years.index.month == month)
    df_match = pd.concat([df_match, df_5_years[mask]])

GH_roof_area = 7.1*4.1 #m^2
df_match['vol_water(m^3)'] = GH_roof_area * (df_match["Total_Precipitation"]/1000)

avg_daily_precip = df_match.groupby(['Year','Month'])['vol_water(m^3)'].mean()
avg_daily_precip

all_avg = avg_daily_precip.mean()
all_avg
print(f'The daily average precipitation of 3 stations is {all_avg:.4f} m^3')
print(f'The daily average precipitation of 3 stations is {all_avg*1000:.1f} Liters')
```
