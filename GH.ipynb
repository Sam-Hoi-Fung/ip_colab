```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('FINAL_DATA.CSV', sep =",", index_col = "MESS_DATUM")
df
```

```python
start = '2018-01-01'
end = '2022-12-31'
df_5_years = df.loc[start:end]
df_5_years
```

```python
if pd.api.types.is_datetime64_any_dtype(df_5_years.index):
    print('yes')
else:
    print('No')
```

```python
pd.options.mode.chained_assignment = None
df_5_years.index = pd.to_datetime(df_5_years.index)
df_5_years['Year'] = df_5_years.index.year
df_5_years['Month'] = df_5_years.index.month
df_5_years
```

```python
df_5_years['Total_Precipitation'] = df_5_years[['1590', '5733', '13670']].sum(axis=1) # Daily sum of three stations
df_5_years.head()
```

## Resampling data to a monthly frequency and then calculating the sum within each month. ##

```python
monthly_accumulated = df_5_years['Total_Precipitation'].resample('M').sum() # In order to get monthly data 
df_5_years['Monthly_Accumulated'] = monthly_accumulated
df_5_years.head(31)
```

## Get the monthly data ##

```python
Monthly = df_5_years.groupby(['Year','Month'])['Monthly_Accumulated'].idxmax()
df_Monthly = df_5_years.loc[Monthly]
df_Monthly.head(24)
```

## Get the highest precipitation value within each year ##

```python
most_precip_month = df_Monthly.groupby(['Year'])['Monthly_Accumulated'].idxmax()
df_most_precip_month = df_Monthly.loc[most_precip_month]
df_most_precip_month
```

## Match the most precipitation month and the 5-year data ##

```python
df_match = pd.DataFrame()
for year, month in zip(df_most_precip_month['Year'], df_most_precip_month['Month']):
    mask = (df_5_years.index.year == year) & (df_5_years.index.month == month)
    df_match = pd.concat([df_match, df_5_years[mask]])
```

```python
df_match.head(62) # A daily data only with the months with the highest precipitaion for each year
```

```python
GH_roof_area = 7.1*4.1 #m^2
df_match['vol_water(m^3)'] = GH_roof_area * (df_match["Total_Precipitation"]/1000)
# divided 1000 to convert millimeters to meters
# the column 'vol_water(m^3)' represents the volume of rain fell on the roof of the Green House
df_match
```

## Daily average precipitation on the green house 

```python
avg_daily_precip = df_match['vol_water(m^3)'].mean()
avg_daily_precip

print(f'The daily average precipitation on the green house is {avg_daily_precip:.4f} m^3 which is equal to {avg_daily_precip*1000:.1f} Liters')
```
